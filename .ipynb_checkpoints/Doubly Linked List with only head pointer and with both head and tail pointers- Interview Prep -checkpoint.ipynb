```python
#DLL Drawbacks as compared to SLL
#It occupies more space as it has two pointers
#You have to adjust more no of pointers while ( Insert, Delete a node )
#As there are more pointers to be adjusted , should be diligent when performing operation


#DLL Operations

class Node:
    
    def __init__(self, info , prev=None, next=None):
        self.info=info
        self.prev=prev
        self.next=next

class LinkedList:
    def __init__(self):
        self.head=None
    def insert_at_begining(self,ele):
        newNode=Node(ele)
        if self.head == None:
            self.head=newNode
        else:
            newNode.next=self.head
            self.head.prev=newNode
            self.head=newNode
    def insert_at_end(self,ele):
        newNode=Node(ele)
        if self.head == None:
            self.head=newNode
        else:
            current=self.head
            while (current.next != None):
                current=current.next
            current.next=newNode
            newNode.prev=current
            

#Delete an Element from a Doubly LINKED LIST

    def delete_node(self,ele):
        if self.head==None:
            print("List is empty")
            return
        #If only one node is present in list
        if self.head.next==None:
            if self.head.info==ele:
                temp=self.head
                self.head=None
                temp=None
                return
            else:
                print("Element is not found in the list")
                return
        
        #delete Node in Between
        temp=self.head
        if temp.info==ele:
            temp.next.prev=None
            self.head=temp.next
            temp=None
            return
        
        temp=self.head.next
        while(temp.next!=None):
            if temp.info==ele:
                temp.prev.next=temp.next
                temp.next.prev=temp.prev
                temp=None
                return
            temp=temp.next
        
        #delete last node present ina list
        
        if temp.info==ele:
            temp.prev.next=None
            temp=None
            return
        print("Element is not found in the list")
        
                
                
    def display(self):
        if self.head == None:
            print("List is empty")
        else:
            current=self.head
            while(current!=None):
                print(current.info)
                current=current.next

ll=LinkedList()
ll.insert_at_begining(10)
ll.insert_at_begining(5)
# ll.insert_at_begining(1000)
ll.insert_at_end(20)
ll.insert_at_end(30)
# ll.insert_at_end(10)
ll.display()
print("$$$$$$$$$$")

ll.delete_node(5)
ll.display()
print("######")
ll.delete_node(30)
ll.display()
print("&‡&&&&&")
ll.delete_node(40)
ll.display()
    
```

```python
#DLL Operation with tail pointer

class tNode:
    
    def __init__(self,info,next=None,prev=None):
        self.info=info
        self.next=next
        self.prev=prev
class dLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        
    def insert_at_begining(self,ele):
        newNode=tNode(ele)
        if self.head==None:
            self.head=newNode
            self.tail=newNode
        else:
            newNode.next=self.head
            self.head.prev=newNode
            self.head=newNode
            
    def insert_at_end(self,ele):
        newNode=tNode(ele)
        if self.head==None:
            self.head=newNode
            self.tail=newNode
        else:
            newNode.prev=self.tail
            self.tail.next=newNode
            self.tail=newNode
            
    def display(self):
        pointer=self.head
        while(pointer!=None):
            print(pointer.info)
            pointer=pointer.next

dll=dLinkedList()
dll.insert_at_begining(10)
dll.insert_at_end(100)
dll.insert_at_end(110)
dll.insert_at_end(150)
dll.insert_at_begining(120)
dll.insert_at_end(210)
dll.insert_at_begining(140)
dll.insert_at_end(190)
dll.display()
            
#140
#120
#10
#100

            
```
